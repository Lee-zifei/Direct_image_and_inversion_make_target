```python
import keras 
from keras.models import  Sequential, load_model, Model 
import matplotlib.pyplot as plt 
from scipy import  io
import numpy as np
```

```python
mat_contents = io.loadmat('../test_data1/X_test.mat')
X_test = mat_contents['X_test'][7740:7742]
print X_test.shape


patch_rows= 32
patch_cols=32
batch_size = 100

filename = 'X_test.dat'
temp = np.fromfile(filename, dtype='float32')
test_datasize = len(temp)/patch_rows/patch_cols
temp = temp.reshape(patch_rows,patch_cols,test_datasize)



X_test = np.zeros([test_datasize,patch_rows,patch_cols])
for i in range(test_datasize):
    X_test[i,:,:] = temp[:,:,i].T
X_test = X_test.reshape(test_datasize, patch_rows, patch_cols, 1)
```

```python
test_datasize, batch_rows, batch_cols = X_test.shape 

X_test = X_test.reshape(test_datasize, batch_rows,batch_cols,1)
print test_datasize, batch_rows, batch_cols
batch_size=40
```

```python
plt.figure
plt.imshow(np.squeeze(X_test[0]))
```

```python
# # load model
model = load_model('../Data/trained_model_smallset.h5')
Y_test = model.predict(X_test, batch_size=batch_size, verbose=1)



io.savemat('./Y_test.mat', {'Y_test':Y_test})
```

```python
layer_names = [layer.name for layer in model.layers]
len(layer_names)
k = 1

for lay in range(1,len(layer_names)):
    interlayer = Model(inputs=model.input, outputs=model.get_layer(layer_names[lay]).output) 
    features = interlayer.predict(X_test) 
    print features[k,:,:,:].shape
    plt.figure()    
    io.savemat('./features%d.mat'%(lay+1), {'features':features[k,:,:,:]})
    if lay < len(layer_names)-1:
        depth = features.shape[3]
        for i in range(depth):
            plt.subplot(4,depth/4,i+1)
            plt.imshow(np.squeeze(features[k,:,:,i]) ,vmin=-0.1, vmax=0.1)
            plt.axis('off')
            plt.subplots_adjust(left=0.1,right=0.9, top=1, bottom=0,wspace=0.01,hspace=.01)
    else:
        plt.figure()
        print features[k,:,:,:].shape
        plt.imshow(np.squeeze(features[k,:,:,:]) ,vmin=-0.1, vmax=0.1)
        
```
