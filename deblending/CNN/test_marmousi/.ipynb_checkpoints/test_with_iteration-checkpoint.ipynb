# load module  

```python
import keras
from keras.models import Sequential, load_model, Model
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
import numpy as np
import matplotlib.pyplot as plt
from keras import backend as K
from scipy import io
import os, sys
from myprog import  *
from scipy import io
```

# set parameters 

```python
n1 = 256 
n2 = 128

patch_rows= 32
patch_cols=32
batch_size = 100
```

# read unblended and blended data 

```python
filename1 = 'hyper.dat'
hyper = np.fromfile(filename1, dtype='float32')
hyper = hyper.reshape(n2,n1).T

filename2 = 'hyper2.dat'
hyper2 = np.fromfile(filename2, dtype='float32')
hyper2 = hyper2.reshape(n2,n1).T


filename3 = 'obser.dat'
X_test1 = np.fromfile(filename3, dtype='float32')
X_test1 = X_test1.reshape(n2,n1).T

filename4 = 'obser2.dat'
X_test2 = np.fromfile(filename4, dtype='float32')
X_test2 = X_test2.reshape(n2,n1).T

filename5 = 'dither.dat'
delay = np.fromfile(filename5, dtype='float32')
```

# The 1st iteration 

```python

def snr(y_true, y_pred):
    return y_pred

reobser1 = myimtocol(X_test1, patch_rows, patch_cols, n1, n2, 1,1, 1)
reobser2 = myimtocol(X_test2, patch_rows, patch_cols, n1, n2, 1,1, 1)
X1_test = np.concatenate((reobser1,reobser2),axis=0)
[test_datasize, patch_rows, patch_cols] = X1_test.shape
X1_test = X1_test.reshape(test_datasize, patch_rows, patch_cols,1)
# load trained model
model = load_model('../Data/trained_model_smallset.h5',custom_objects={'snr':snr})
# prediction
Y1_test = model.predict(X1_test, batch_size=batch_size, verbose=1)

Y1_test = np.squeeze(Y1_test)

[nn1, nn2, nn3]  = Y1_test.shape
d1de1 = myimtocol(Y1_test[0:nn1/2,:,:], patch_rows, patch_cols, n1, n2, 1,1, 0)
d2de1 = myimtocol(Y1_test[nn1/2:nn1,:,:], patch_rows, patch_cols, n1, n2, 1,1, 0)

plt.figure()
plt.subplot(121);plt.imshow(d1de1)
plt.subplot(122);plt.imshow(d2de1)
```

# The 2nd iteration 

```python
temp1 = X_test1 - dither (d2de1, delay);
temp2 = X_test2 - dither (d1de1, -delay);
reobser3 = myimtocol(temp1, patch_rows, patch_cols, n1, n2, 1,1, 1)
reobser4 = myimtocol(temp2, patch_rows, patch_cols, n1, n2, 1,1, 1)
X2_test = np.concatenate((reobser3,reobser4),axis=0)
X2_test = X2_test.reshape(test_datasize, patch_rows, patch_cols,1)
Y2_test = model.predict(X2_test, batch_size=batch_size, verbose=1)
Y2_test = np.squeeze(Y2_test)

d1de2 = myimtocol(Y2_test[0:nn1/2,:,:], patch_rows, patch_cols, n1, n2, 1,1, 0)
d2de2 = myimtocol(Y2_test[nn1/2:nn1,:,:], patch_rows, patch_cols, n1, n2, 1,1, 0)

plt.figure()
plt.subplot(121);plt.imshow(d1de2)
plt.subplot(122);plt.imshow(d2de2)
```

# The 3rd iteration 

```python
temp3 = X_test1 - dither (d2de2, delay);
temp4 = X_test2 - dither (d1de2, -delay);
reobser5 = myimtocol(temp3, patch_rows, patch_cols, n1, n2, 1,1, 1)
reobser6 = myimtocol(temp4, patch_rows, patch_cols, n1, n2, 1,1, 1)
X3_test = np.concatenate((reobser5,reobser6),axis=0)
X3_test = X3_test.reshape(test_datasize, patch_rows, patch_cols,1)
Y3_test = model.predict(X3_test, batch_size=batch_size, verbose=1)
Y3_test = np.squeeze(Y3_test)

d1de3 = myimtocol(Y3_test[0:nn1/2,:,:], patch_rows, patch_cols, n1, n2, 1,1, 0)
d2de3 = myimtocol(Y3_test[nn1/2:nn1,:,:], patch_rows, patch_cols, n1, n2, 1,1, 0)

plt.figure()
plt.subplot(121);plt.imshow(d1de3)
plt.subplot(122);plt.imshow(d2de3)
```
